```python
# Cell 1: Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Establecer estilo de los gráficos
sns.set(style="whitegrid")
```

```python
# Cell 2: Cargar los datos
file_path = '../data/raw/accidentes_transito.csv'
df = pd.read_csv(file_path)
```

```python
# Mostrar las primeras filas del dataframe para una vista general
df.head()
```

```python
# Cell 3: Resumen de los datos
# Mostrar la cantidad de filas y columnas
print(f"El conjunto de datos tiene {df.shape[0]} filas y {df.shape[1]} columnas.")
```

```python
# Información sobre el tipo de datos
df.info()
```

```python
# Verificar la cantidad de valores nulos por columna
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# Cell 4: Descripción estadística de las variables numéricas
df.describe()
```

```python
# Cell 4: Descripción estadística de las variables numéricas
df.describe()
```

```python
# Cell 5: Visualización de la distribución de los accidentes por distrito
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Nom_districte', order=df['Nom_districte'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribución de accidentes por distrito')
plt.show()
```

```python
# Cell 6: Visualización de la relación entre día de la semana y cantidad de accidentes
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Descripcio_dia_setmana', order=df['Descripcio_dia_setmana'].value_counts().index)
plt.title('Distribución de accidentes según el día de la semana')
plt.show()
```

```python
# Cell 7: Distribución de accidentes por causa mediata
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Descripcio_causa_mediata', order=df['Descripcio_causa_mediata'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribución de causas de accidentes')
plt.show()
```

```python
# Cell 8: Relación entre la hora del día y la cantidad de accidentes
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Hora_dia', bins=24, kde=False)
plt.title('Accidentes por hora del día')
plt.show()
```

```python
# Cell 9: Relación entre la latitud y la longitud (mapa de dispersión de accidentes)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Longitud_WGS84', y='Latitud_WGS84')
plt.title('Distribución geográfica de accidentes')
plt.show()
```
